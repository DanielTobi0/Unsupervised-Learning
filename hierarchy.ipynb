```python
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
method_complete = linkage(X, method='complete', metric='euclidean') # method: average, single
dendrogram(method_complete)
plt.show()
```

```python
cluster_labels = cut_tree(method_complete, n_clusters=3)
```

```python
cols = ['Petal Length', 'Petal Width', 'Sepal Length', 'Sepal Width']
data = pd.DataFrame(X, columns=cols)
data['cluster_label'] = cluster_labels
```
